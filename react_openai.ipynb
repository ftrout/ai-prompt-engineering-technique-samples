```python
import openai
import os
from dotenv import load_dotenv

load_dotenv()
openai.api_key = os.getenv("OPENAI_API_KEY")
if not openai.api_key:
    openai.api_key = input("Paste your OpenAI API key: ").strip()
print("OpenAI ready!")
```

## ReAct System Prompt Example

```python
SYSTEM_PROMPT = """
You are an agent using ReAct: Reason about the task, Act (suggest a tool/query), Observe (simulate result), repeat until solved.
Format: Thought: [reasoning]
Action: [action]
Observation: [simulated]
Final Answer: [solution]
"""
```

## Interactive ReAct Agent

```python
messages = [{"role": "system", "content": SYSTEM_PROMPT}]

print("ReAct Agent Ready! Enter a task. Type 'quit' to exit.\n")

while True:
    user_input = input("Your task: ")
    if user_input.lower() in ["quit", "exit"]:
        print("Goodbye!")
        break
    messages.append({"role": "user", "content": user_input})
    response = openai.chat.completions.create(
        model="gpt-4o-mini",
        messages=messages,
        temperature=0.3
    )
    reply = response.choices[0].message.content
    print(f"\nProcess:\n{reply}\n")
    messages.append({"role": "assistant", "content": reply})
```
