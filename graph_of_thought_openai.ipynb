````
import openai, os, json, networkx as nx
from dotenv import load_dotenv
load_dotenv()
openai.api_key = os.getenv("OPENAI_API_KEY") or input("OpenAI key: ")

def graph_of_thought(problem):
    print("Building reasoning graph...\n")
    
    got = openai.chat.completions.create(
        model="gpt-4o-mini",
        messages=[{"role": "user", "content": f"""
Solve this using Graph-of-Thought reasoning:
{problem}

Output exactly in this JSON format:
{{
  "nodes": {{"1": "thought text", "2": "..."}},
  "edges": [[1,2,"supports"],[2,3,"contradicts"],[1,4,"expands"]],
  "final_answer": "the solution"
}}
Use merging, looping, or transformation if helpful.
"""}],
        temperature=0.8
    ).choices[0].message.content

    try:
        data = json.loads(got.strip("```json").strip("```"))
        print("Graph generated!\n")
        for k,v in data["nodes"].items():
            print(f"Node {k}: {v[:120]}{'...' if len(v)>120 else ''}")
        print(f"\nFinal Answer: {data['final_answer']}")
        return data
    except:
        print("Raw output:\n", got)
        return got
````

````
problem = input("Complex problem: ")
graph_of_thought(problem)
````
