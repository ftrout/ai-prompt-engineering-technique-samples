```python
import openai
import os
from dotenv import load_dotenv

load_dotenv()
openai.api_key = os.getenv("OPENAI_API_KEY")
if not openai.api_key:
    openai.api_key = input("Paste your OpenAI API key: ").strip()
print("OpenAI ready!")
```

## Play Off System Prompt Example

```python
SYSTEM_PROMPT = """
You are a creative ideation expert using the play off method.
Start with a base idea.
Then, play off it by contrasting (e.g., opposite twist), combining, or iterating 3-5 times.
End with a polished final concept.
"""
```

## Interactive Play Off Generator

```python
messages = [{"role": "system", "content": SYSTEM_PROMPT}]

print("Play Off Idea Generator Ready! Enter a starting idea. Type 'quit' to exit.\n")

while True:
    user_input = input("Your idea: ")
    if user_input.lower() in ["quit", "exit"]:
        print("Goodbye!")
        break
    messages.append({"role": "user", "content": user_input})
    response = openai.chat.completions.create(
        model="gpt-4o-mini",
        messages=messages,
        temperature=0.8  # Higher for creativity
    )
    reply = response.choices[0].message.content
    print(f"\nRefined Ideas:\n{reply}\n")
    messages.append({"role": "assistant", "content": reply})
```
