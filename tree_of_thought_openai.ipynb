```python
import openai
import os
from dotenv import load_dotenv

load_dotenv()
openai.api_key = os.getenv("OPENAI_API_KEY")
if not openai.api_key:
    openai.api_key = input("Paste your OpenAI API key: ").strip()
print("OpenAI ready!")
```

## ToT System Prompt Example

```python
SYSTEM_PROMPT = """
You are a decision-making expert using tree of thought.
For queries, generate 3-5 branches of reasoning.
For each branch: Describe it, list pros/cons, score 1-10.
Then, evaluate all and pick the best with final recommendation.
"""
```

## Interactive ToT Analyzer

```python
messages = [{"role": "system", "content": SYSTEM_PROMPT}]

print("Tree of Thought Analyzer Ready! Enter a decision/query. Type 'quit' to exit.\n")

while True:
    user_input = input("Your query: ")
    if user_input.lower() in ["quit", "exit"]:
        print("Goodbye!")
        break
    messages.append({"role": "user", "content": user_input})
    response = openai.chat.completions.create(
        model="gpt-4o-mini",
        messages=messages,
        temperature=0.5
    )
    reply = response.choices[0].message.content
    print(f"\nAnalysis:\n{reply}\n")
    messages.append({"role": "assistant", "content": reply})
```
